```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
import warnings

warnings.filterwarnings("ignore")

path_to_bots = "/Users/jiazhenzhang/Desktop/Machine_learning"
combined_data = pd.read_csv(path_to_bots+"/training_data_2_csv_UTF.csv")

ratio_column = []
result_column = []
idlen_column = []
bot_in_description_column = []
screen_name_column = []
len_name = []
for index, row in combined_data.iterrows():
    followers = row['followers_count']
    friends = row['friends_count']
    listcount = row['listedcount']
    len_name.append(len(str(row['name'])))
    ratio = 0
    if followers == 0:
        ratio = friends * 100
    else:
        ratio = float(friends) / float(followers)
    ratio_column.append(ratio)
    idlen_column.append(len(str(row['id'])))
    
    descript_str = str(row['description']).lower()
    if len(descript_str) == 0:
        bot_in_description_column.append(2)
    elif "bot" in descript_str:
        bot_in_description_column.append(1)
    else:
        bot_in_description_column.append(0)
    
    screen_name_str = str(row['screen_name']).lower()
    if "bot" in screen_name_str:
        screen_name_column.append(True)
    else:
        screen_name_column.append(False)

count_vect = CountVectorizer()
combined_data['description'].fillna("#$%^&*@", inplace=True)
combined_data_filled = combined_data

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
fourth_data = combined_data[['default_profile', 'default_profile_image', 'verified', 'has_extended_profile', 'friends_count', 'followers_count', 'listedcount', 'statuses_count', 'favourites_count']]
fourth_data['ratio'] = ratio_column
fourth_data['name'] = len_name
fourth_data['idlenth'] = idlen_column
fourth_data['bot_in_description'] = bot_in_description_column
fourth_data['bot_in_screenname'] = screen_name_column
#print(type(fourth_data['default_profile'][0]), type(fourth_data['default_profile_image'][0]),type(fourth_data['verified'][0]),type(fourth_data['has_extended_profile'][0]), type(fourth_data['friends_count'][0]), type(fourth_data['followers_count'][0]), type(fourth_data['listedcount'][0]),type(fourth_data['statuses_count'][0]), type(fourth_data['favourites_count'][0]),type(fourth_data['ratio'][0]), type(fourth_data['name'][0]))
fourth_data.head(5)
```
